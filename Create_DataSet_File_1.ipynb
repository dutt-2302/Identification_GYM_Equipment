define classes that contain data

```
classes=['treadmill','dumb_bell','bench_press','elliptical','multi_machine','rowing_machine','aerobic_steppers']
```

define filess that contains file name


```
filess=['treadmill.csv','dumb_bell.csv','bench_press.csv','elliptical.csv','multi_machine.csv','rowing_machine.csv','aerobic_steppers.csv']
```

downloading data

```
for i in range(0,len(classes)):
    folder = classes[i]
    url_file = filess[i]
    path = Path('gym_data')
    dest = path/folder
    dest.mkdir(parents=True, exist_ok=True)
    download_images(url_file, dest, max_pics=1000)
```

deleting non readable image

```
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```
