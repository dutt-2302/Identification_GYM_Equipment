**taking random image from the google for making the vedio**

```
!pip install bing-image-downloader
from bing_image_downloader import downloader
gym_tools = ['treadmill','dumb_bell','bench_press','elliptical','multi_machine','rowing_machine','aerobic_steppers']
for gym in gym_tools:
  downloader.download(gym, limit=10,  output_dir='/test_data1', adult_filter_off=True, force_replace=False, timeout=60)
```

**shuffling the image**

```
all_images=[]
for gym in gym_tools: 
  images = os.listdir("/content/test_data1/"+ gym)
  for image in images:
    im = "/content/test_data1/"+ gym +"/"+image
    all_images.append(im)
import random
random.shuffle(all_images)
```

**making vedio**

```
import cv2
os.makedirs("test_result",exist_ok=True)
i=0
for test_file in all_images:
  img=cv2.imread(test_file)
  cv2.imwrite("temp.jpg",img)
  ans_text=my_predictor("temp.jpg")
  font = cv2.FONT_HERSHEY_SIMPLEX 
  i+=1
  org = (0, 50) 
  fontScale = 2
  color = (255, 0, 0) 
  thickness = 4
  img=cv2.resize(img, (600, 600))
  img = cv2.putText(img, str(ans_text), org, font,fontScale, color, thickness, cv2.LINE_AA) 
  cv2.imwrite("test_result/"+str(i)+".jpg", img)
```
